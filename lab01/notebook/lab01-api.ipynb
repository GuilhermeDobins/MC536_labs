```python
#Os primeiros dados são xml, retirados da database da camara legislativa
import io
import requests
import xml.etree.ElementTree as et
url="https://www.camara.leg.br/sitcamaraws/deputados.asmx/ObterDeputados"
data = requests.get(url).content
x = et.parse(io.StringIO(data.decode("utf-8")))
print(x.find("deputado/ideCadastro").text)
```
